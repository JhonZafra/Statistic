### Uso de la distribución binomial

Genera un número aleatorio correspondiente al número de éxitos en $n$ intentos, donde la probabilidad de éxito de un intento es $p$

```python
#np.random.seed(123)
np.random.binomial(n = 30, p = 0.5)
```

```python
caras = []
L = 10**5

y = np.random.binomial(n = 30, p = 0.5, size = 10**5)

for k in range(L):
    cara = 0
    for n in range(30):
        r = np.random.rand()
        if r < 0.5:
            cara += 1
    caras.append(cara)
print(caras)


plt.figure(figsize= (15,5))
cuentas2, ejes2, barras2 = plt.hist(caras, bins= 31, 
                                    edgecolor= 'white', range= (0, 31), alpha = 0.5,
                                    label = 'Experimento\nmonedas')

plt.hist(y, bins= 31, 
        edgecolor= 'white', range= (0, 31), alpha = 0.5,
        label = 'Experimento\nnumpy' )

#sns.kdeplot(caras, bw_adjust= 3) # La curva es más suave para mayores valores de bw_adjust
plt.bar_label(barras2)
plt.title('Probabilidades de resultados para 30 lanzamientos', size= 16, weight= 'bold')
plt.xlabel('Número de caras')
plt.ylabel('Probabilidad')
plt.xlim(5, 25)
plt.xticks(range( 5, 25 ))
plt.legend()
plt.show()
```

### PROBABILIDAD DE LA DISTRIBUCIÓN BINOMIAL

Clase 22/06/2023

Probabilidad de tener k éxitos en n intentos cuando la probabilidad de éxito en un intento es p.

$$ P(k,n,p) = \begin{pmatrix} n \\ k\end{pmatrix} p ^k (1-p)^{n-k} $$

$$ \begin{pmatrix} n\\k \end{pmatrix} = \frac{n!}{k! (n-k)!}$$

Definir la función factorial

$$ n! = n(n-1)(n-2) \cdots (2)(1) $$

$$ 5! = (5)(4)(3)(2)(1) $$

$$ 1! = 1 $$

$$ 0! = 1  $$

```python
def factorial(n):
    x = 1
    if n > 1:
        for k in range (2, n+1):
            x *= k
    return x
```

```python
factorial(5)
```

```python
n = 5
list(range(2 , n+1))
```

$$ \begin{pmatrix} n\\k \end{pmatrix} = \frac{n!}{k! (n-k)!}$$

```python
def CB(n, k):
    if n >= k:
        return factorial(n) / (factorial(k) * factorial(n-k))
```

```python
print(CB(6,6))
print(CB(6,5))
print(CB(6,4))
print(CB(6,3))
print(CB(6,2))
print(CB(6,1))
print(CB(6,0))
```
