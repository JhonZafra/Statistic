```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10**4
np.random.seed(123)
d1 = np.random.choice([1, 2, 3, 4, 5, 6], N)
d2 = np.random.choice([1, 2, 3, 4, 5, 6], N)
```

```python
plt.figure(figsize= (7,4), facecolor= "lightblue")
cuentas, ejes , barras = plt.hist(d1 + d2,
         edgecolor="white",
         color='brown',
         density= True, # Permite calular las fracciones respecto al total de datos y ponerlas en la gráfica
         bins= 11, 
         range = (2, 13))
plt.bar_label(barras, fmt='%.3f')
#for n in range(len(cuentas)):
    #plt.text(ejes[n] + 0.15, cuentas[n]/2, round(cuentas[n]), color = 'white')
#plt.ylim(0, 1850)
plt.xticks(np.arange(2,13)+ 0.5, np.arange(2,13), color= 'green')
plt.title("Lanzamiento de dos dados", weight= 'bold', size= 16)
plt.xlabel('Resultado de la suma', color= 'purple', size=11)
plt.ylabel('Frecuencia de cada resultado', color= 'purple', size=11)
plt.show()
```

##### Distribución uniforme

* np.random.rand() genera números uniformemente distribuidos entre 0 y 1.

```python
N = 10**5
np.random.seed(123)
x = np.random.rand(N)
```

```python
plt.figure(figsize= (7,5) )
cuentas, ejes, barras = plt.hist(x, bins= 2, edgecolor= 'white')
plt.bar_label(barras)
plt.show()
```

```python
r = np.random.rand() # Genera números aleatorios entre 0 y 1.

if r < 0.5:
    print('Cara')
else:
    print('Sello')
```

```python
cara = 0
sello = 0

for n in range(30):
    r = np.random.rand()
    if r < 0.5:
        cara += 1
    else:
        sello += 1

print(f'Caras = {cara} \nSellos = {sello}')
plt.bar(['Caras', 'Sellos'], [cara, sello])
plt.show()
```
