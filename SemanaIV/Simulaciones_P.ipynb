```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10**4
np.random.seed(123)
d1 = np.random.choice([1, 2, 3, 4, 5, 6], N)
d2 = np.random.choice([1, 2, 3, 4, 5, 6], N)
```

```python
plt.figure(figsize= (7,4), facecolor= "lightblue")
cuentas, ejes , barras = plt.hist(d1 + d2,
         edgecolor="white",
         color='brown',
         density= True, # Permite calular las fracciones respecto al total de datos y ponerlas en la gráfica
         bins= 11, 
         range = (2, 13))
plt.bar_label(barras, fmt='%.3f')
#for n in range(len(cuentas)):
    #plt.text(ejes[n] + 0.15, cuentas[n]/2, round(cuentas[n]), color = 'white')
#plt.ylim(0, 1850)
plt.xticks(np.arange(2,13)+ 0.5, np.arange(2,13), color= 'green')
plt.title("Lanzamiento de dos dados", weight= 'bold', size= 16)
plt.xlabel('Resultado de la suma', color= 'purple', size=11)
plt.ylabel('Frecuencia de cada resultado', color= 'purple', size=11)
plt.show()
```

##### Distribución uniforme

* np.random.rand() genera números uniformemente distribuidos entre 0 y 1.

```python
N = 10**5
#np.random.seed(123)
```

```python
x = np.random.rand(N)
plt.figure(figsize= (7,5) )
cuentas, ejes, barras = plt.hist(x, bins= 2, edgecolor= 'white')
plt.bar_label(barras)
plt.text(0.2, cuentas[0]/2, round(cuentas[0] / N, 4), color= 'white')
plt.text(0.7, cuentas[1]/2, round(cuentas[1] / N, 4), color= 'white')
plt.show()
```

```python
r = np.random.rand() # Genera números aleatorios entre 0 y 1.
print(r)
if r < 0.5:
    print('Cara')
else:
    print('Sello')
```

```python
cara = 0
sello = 0

for n in range(30):
    r = np.random.rand()
    if r < 0.5:
        cara += 1
    else:
        sello += 1

print(f'Caras = {cara} \nSellos = {sello}')
plt.bar(['Caras', 'Sellos'], [cara, sello])
plt.text(0 - 0.1, cara/2, cara, color = 'white', size= 15)
plt.text(1 - 0.1, sello/2, sello, color = 'white', size= 15)
plt.show()
```

¿Cuél es la probabilidad de obtener 15 caras en 30 lanzamientos?

```python
import seaborn as sns
```

```python
caras = []
L = 10**5

for k in range(L):
    cara = 0
    for n in range(30):
        r = np.random.rand()
        if r < 0.5:
            cara += 1
    caras.append(cara)
print(caras)


plt.figure(figsize= (15,5))
cuentas2, ejes2, barras2 = plt.hist(caras, bins= 31, 
                                    density = True,
                                    edgecolor= 'white', range= (0, 31))
sns.kdeplot(caras, bw_adjust= 3) # La curva es más suave para mayores valores de bw_adjust
plt.bar_label(barras2, fmt= '%.3f')
plt.title('Probabilidades de resultados para 30 lanzamientos', size= 16, weight= 'bold')
plt.xlabel('Número de caras')
plt.ylabel('Probabilidad')
plt.xlim(5, 25)
plt.xticks(range( 5, 25 ))
plt.show()
```

### Uso de la distribución binomial

Genera un número aleatorio correspondiente al número de éxitos en $n$ intentos, donde la probabilidad de éxito de un intento es $p$

```python
#np.random.seed(123)
np.random.binomial(n = 30, p = 0.5)
```

```python
caras = []
L = 10**5

y = np.random.binomial(n = 30, p = 0.5, size = 10**5)

for k in range(L):
    cara = 0
    for n in range(30):
        r = np.random.rand()
        if r < 0.5:
            cara += 1
    caras.append(cara)
print(caras)


plt.figure(figsize= (15,5))
cuentas2, ejes2, barras2 = plt.hist(caras, bins= 31, 
                                    edgecolor= 'white', range= (0, 31), alpha = 0.5,
                                    label = 'Experimento\nmonedas')

plt.hist(y, bins= 31, 
        edgecolor= 'white', range= (0, 31), alpha = 0.5,
        label = 'Experimento\nnumpy' )

#sns.kdeplot(caras, bw_adjust= 3) # La curva es más suave para mayores valores de bw_adjust
plt.bar_label(barras2)
plt.title('Probabilidades de resultados para 30 lanzamientos', size= 16, weight= 'bold')
plt.xlabel('Número de caras')
plt.ylabel('Probabilidad')
plt.xlim(5, 25)
plt.xticks(range( 5, 25 ))
plt.legend()
plt.show()
```
