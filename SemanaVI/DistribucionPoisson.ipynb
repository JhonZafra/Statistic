*Ejemplo:*

Simular la llegada de clientes a una sucursal donde llega un cliente cada dos(2) minutos. $ \lambda = \frac {1}{2} = \frac {1  \text{cliente}}{2 \text{ minutos}} = 0.5 $. Voy a hacer una simulaci貌n durante 60 minutos.

Comparo el resultado con el que se obtiene al generar 10000 tiempos para llegadas de clientes siguiendo una distribucion exponencial.

```python
位 = 0.5 # N煤mero de clientes por minuto
tiempo_total = 60

np.random.seed(100)
intervalos_tiempo = np.random.exponential(scale= 1/位, size= 10**4) # Tiempo promedio entre dos eventos suscesivos
tiempo_llegada = np.cumsum(intervalos_tiempo) # cumsum genera un arreglo con una suma acumulada
```

```python
intervalos_tiempo.mean()
```

```python
plt.hist(intervalos_tiempo, bins=100, density= True, label='Histograma \nsimulaci贸n')
plt.plot(t, y, color='red', label='Distribuci贸n \nexponencial')
plt.legend()
plt.show()
```

```python
plt.plot(range(10**4), tiempo_llegada)
plt.xlabel('N煤mero de clientes')
plt.ylabel('Tiempo de llegada \n[min]')
plt.show()
```

Ahora vamos a simular la llegada de clientes durante una hora ( 60 minutos)

```python
位 = 0.5 
tiempo_total = 60

np.random.seed(100)
intervalos_tiempo = [np.random.exponential(scale= 1/位)] 

while np.array(intervalos_tiempo).sum() <= 60:
    intervalos_tiempo.append(np.random.exponential(scale= 1/位))

tiempo_llegada = np.cumsum(intervalos_tiempo)
```

```python
plt.hist(intervalos_tiempo, density= True, label= 'Histograma \nSimulaci贸n')
plt.plot(t, y, color='red', label= 'Distribuci贸n \nExponencial')
plt.legend()
plt.show()
```

```python
plt.step(tiempo_llegada, np.arange(1, len(tiempo_llegada) + 1))
plt.xlabel('Tiempo de llegada \n[min]')
plt.ylabel('N煤mero de clientes')
plt.show()
```
