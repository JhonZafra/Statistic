```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cálculo de Pi sin interaciones
def piMC(N):
    x = np.random.rand(N)
    y = np.random.rand(N)
    c = 0

    for n in range(N):
        if x[n]**2 + y[n]**2 < 1:
            c += 1
    return 4*c/N

piMC(10**3)
```

Ecuación de la circunferencia
$$ x^2 + y^2  = 1$$
$$ y  = \sqrt {1 - x^2}$$


```python
N = 1000
x = np.random.rand(N)
y = np.random.rand(N)

t = np.linspace(0,1)
s = np.sqrt(1 -t**2)

plt.figure(figsize= (3,3))
plt.scatter(x, y, s= 2) # s hace referencia a size
plt.plot( t, s, color = 'red')
plt.show()
```

```python
# Cálculo de pi con iteraciones

def IteraPiMC(iteraciones, n_puntos):
    Pi = []
    for n in range(iteraciones):
        Pi.append(piMC(n_puntos))
    print(f'El promedio es :{np.mean(Pi)}')
    print(f'La desviación es :{np.std(Pi)}')
    plt.hist(Pi, bins=20)
    plt.axvline(np.mean(Pi), color = 'red', label= f'$\pi$ promedio = {np.mean(Pi).round(4)}')
    plt.legend()
    plt.title('Aproximaciones de $\pi$')
    plt.show()
    

IteraPiMC(10**4, 10**3)
```

Verificaciones del teorema del límite central

Distribución uniforme

```python
N = 10**4
x = 10 + 10*np.random.rand(N)
print(f'Promedio = {x.mean().round(3)}')

plt.figure(figsize= (4,3))
plt.hist(x)
plt.axvline(x.mean(), color='red')
plt.show()
```

```python
N = 10**4 # Cantidad de números generados
M = 10**4 # Número de iteraciones
z = []
for _ in range(M):   
    x = 10 + 10*np.random.rand(N)
    z.append(x.mean())
print(f'Promedio = {np.array(z).mean().round(4)}')
```

```python
z = np.array(z)
plt.figure(figsize= (4,3))
plt.hist(z, bins= 100)
plt.axvline(z.mean(), color='red')
plt.xlabel('Promedios')
plt.ylabel('Iteraciones')
plt.show()
```

#### *Ejercicio:* Calculo de la integral

$$ \int_0^\pi  \sin(x)dx = 2$$

```python
def integral(N):
    x = np.random.rand(N) * np.pi
    y = np.random.rand(N) 
    c = 0

    for n in range(N):
        if y[n] < np.sin(x[n]) :
            c += 1
    return (c/N)*np.pi

integral(10**3)
```

```python
N = 1000
x = np.random.rand(N) * np.pi
y = np.random.rand(N) 

t = np.linspace(0, np.pi)
s = np.sin(t)

plt.figure(figsize= (4,3))
plt.plot(t, s)
plt.xticks([0, np.pi], [0, 'π'])
plt.scatter(x, y, s = 5, color= 'red')
plt.show()
```

```python
N = 10**3
M = 10**4
A = []

for _ in range(M):
    x = np.random.rand(N) * np.pi
    y = np.sin(x)
    A.append(np.mean(y) * np.pi)

print(f'El promedio es: {np.mean(A)}')
print(f'La desviación es: {np.std(A)}')
```
