```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def piMC(N):
    x = np.random.rand(N)
    y = np.random.rand(N)
    c = 0
    for n in range(N):
        if x[n]**2 + y[n]**2 < 1:
            c += 1
    return 4*c/N

piMC(10**3)
```

```python
def IteraPiMC(iteraciones, n_puntos):
    Pi = []
    for n in range(iteraciones):
        Pi.append(piMC(n_puntos))
    print(f'El promedio es :{np.mean(Pi)}')
    print(f'La desviaciÃ³n es :{np.std(Pi)}')
    plt.hist(Pi)
    plt.title('Aproximaciones de $\pi$')
    plt.show()

IteraPiMC(10**4, 10**3)
```
